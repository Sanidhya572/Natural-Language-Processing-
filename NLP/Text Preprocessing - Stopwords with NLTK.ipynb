```python
paragraph = """ 
My dear citizens,

As we stand on the threshold of the future, let us remember that our actions today shape the world of tomorrow. Each one of us holds within us the power to create positive change, to build a nation that thrives on knowledge, innovation, and compassion.

Let us embrace the spirit of unity and diversity that defines our great nation. In diversity, we find strength, resilience, and the richness of human experience. Let us celebrate our differences and work together towards a common vision of progress and prosperity for all.

Education is the key that unlocks the doors of opportunity. Let us strive to provide quality education to every child, empowering them to dream, to aspire, and to achieve greatness. For it is through education that we can eradicate ignorance, poverty, and inequality.

I urge the youth of our nation to dream big, to set audacious goals, and to never lose sight of their potential to make a difference. Remember, the journey to success is paved with hard work, perseverance, and a relentless pursuit of excellence.

Let us also be mindful of our responsibilities towards the environment. We are the custodians of this planet, and it is our duty to protect and preserve it for future generations.

In closing, I call upon each and every one of you to join hands in building a brighter tomorrow. Together, let us dare to dream, to innovate, and to create a world where peace, prosperity, and progress reign supreme.

Jai Hind!
"""
```

```python
from nltk.stem import PorterStemmer
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
stopwords.words('german')
```

```python
stopwords.words('arabic')
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
# Apply stopwords and Filter and then apply stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
    
```

```python
sentences
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
# Apply stopwords and Filter and then apply snowball stemming
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
    
```

```python
sentences
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(sentences)):
    #sentences[i]=sentences[i].lower()
    words=nltk.word_tokenize(sentences[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```
